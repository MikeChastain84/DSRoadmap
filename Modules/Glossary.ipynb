## Load Glossary

```python
import numpy as np
import pandas as pd
import re

def convert_to_array(x):
    """Converts a string representation of a list to a NumPy array."""
    x = re.sub(r'[\[\]]', '', x)
    x = x.split()
    x = np.array(x, dtype=float)
    return x

glossary = pd.read_csv('https://raw.githubusercontent.com/gitmystuff/DSRoadmap/refs/heads/main/Modules/glossary_with_embeddings.csv')
glossary['encoded'] = glossary['encoded'].apply(convert_to_array)
glossary.head(10)
```

```python
import pprint
import textwrap

terms = glossary[glossary['combined'].str.lower().str.contains('polynomial')]

for index, row in terms.iterrows():
    row_dict = row.to_dict()
    print("\033[1mModule:\033[0m", row_dict['module'])  # \033[1m for bold, \033[0m for reset
    print("\033[1mTerm:\033[0m", row_dict['term'])
    print("\033[1mDefinition:\033[0m", textwrap.fill(row_dict['definition'], width=60))
    print("\n")
```

```python
import requests
import numpy as np
from transformers import pipeline, AutoTokenizer
from google.colab import userdata
import textwrap

model_id = 'sentence-transformers/all-MiniLM-L6-v2'
api_embedding = f'https://api-inference.huggingface.co/pipeline/feature-extraction/{model_id}'
headers = {'Authorization': f'Bearer {userdata.get("HF_TOKEN")}'}

def wrap(x):
  return textwrap.fill(x, replace_whitespace=False, fix_sentence_endings=True)

def vector_similarity(vec1, vec2):
  try:
    return np.dot(np.array(vec1), np.array(vec2))
  except:
    return 0 # This ensures that the similarity column will be numeric

def embed(texts):
    response = requests.post(api_embedding, headers=headers, json={"inputs": texts})
    embeddings = response.json()
    return embeddings

def term_query():
  question = input('Enter a term: ')
  embedded_prompt = embed(question)
  glossary['similarity'] = glossary['encoded'].apply(lambda vector: vector_similarity(vector, embedded_prompt))
  glossary['similarity'] = pd.to_numeric(glossary['similarity'], errors='coerce')
  return None
```

```python
term_query()
glossary.nlargest(5, 'similarity') # returns n = 5 most similar rows
```
